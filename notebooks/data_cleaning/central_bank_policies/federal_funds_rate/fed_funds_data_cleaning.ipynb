```python
fed_funds = pd.read_excel(f"../../../{FED_FUNDS_PATH}FEDFUNDS.xls")
fed_funds = fed_funds[10:]
fed_funds
```

```python
col_names = {
    "FRED Graph Observations":"Date",
    "Unnamed: 1": "Fed_Fund_rate"
}
```

```python
fed_funds = fed_funds.rename(columns=col_names)
fed_funds
```

```python
fed_funds["Date"] = pd.to_datetime(fed_funds["Date"])
fed_funds
```

```python
fed_funds["Year"] = fed_funds['Date'].dt.year
fed_funds["Month"] = fed_funds['Date'].dt.month
fed_funds["Day"] = fed_funds["Date"].dt.day
fed_funds
```

```python
fed_funds = fed_funds.drop("Date", axis=1)
fed_funds
```

```python
bins   = [0, 3, 6, 8, 12]
labels = ["Q1", "Q2", "Q3", "Q4"]
fed_funds["Quarter"] = pd.cut(fed_funds["Month"], bins=bins, labels=labels, right=True)
fed_funds
```
