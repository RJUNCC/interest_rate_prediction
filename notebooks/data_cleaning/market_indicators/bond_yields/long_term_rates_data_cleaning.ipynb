```python
long_term_rates = pd.read_csv(f"../../../{CSV_DATA_PATH}long_term_rates.csv")
long_term_rates
```

```python
long_term_rates["Date"]  = pd.to_datetime(long_term_rates["Date"], format="%m/%d/%y")
long_term_rates["Year"]  = long_term_rates["Date"].dt.year
long_term_rates["Month"] = long_term_rates["Date"].dt.month
long_term_rates["Day"]   = long_term_rates["Date"].dt.day
long_term_rates = long_term_rates.drop("Date", axis=1)
```

```python
long_term_rates = long_term_rates.drop("Extrapolation Factor", axis=1)
```

```python
long_term_rates
```

```python
bins   = [0, 3, 6, 8, 12]
labels = ["Q1", "Q2", "Q3", "Q4"]
long_term_rates["Quarter"] = pd.cut(long_term_rates["Month"], bins=bins, labels=labels, right=True)
long_term_rates
```

```python
path = Path(Path.cwd()/"../../../../data")
path
```

```python
data_folder = path/'parquet/bonds_rates/long_term_rates'
data_folder.mkdir(parents=True, exist_ok=True)
```

```python
long_term_rates.to_parquet(data_folder/"long_term_rates.parquet")
```
