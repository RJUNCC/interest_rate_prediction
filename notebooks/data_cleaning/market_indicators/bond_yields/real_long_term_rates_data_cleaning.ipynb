```python
real_long_term = pd.read_csv(f"../../../{CSV_DATA_PATH}real_long_term_rates.csv")
real_long_term
```

```python
real_long_term["Date"]  = pd.to_datetime(real_long_term["Date"], format="%m/%d/%y")
real_long_term["Year"]  = real_long_term["Date"].dt.year
real_long_term["Month"] = real_long_term["Date"].dt.month
real_long_term["Day"]   = real_long_term["Date"].dt.day
real_long_term = real_long_term.drop("Date", axis=1)
```

```python
real_long_term
```

```python
bins   = [0, 3, 6, 8, 12]
labels = ["Q1", "Q2", "Q3", "Q4"]
real_long_term["Quarter"] = pd.cut(real_long_term["Month"], bins=bins, labels=labels, right=True)
real_long_term
```
