```python
yield_curve_rates = pd.read_csv(f"../../../{CSV_DATA_PATH}yield_curve_rates.csv")
yield_curve_rates
```

```python
yield_curve_rates["Date"]  = pd.to_datetime(yield_curve_rates["Date"], format="%m/%d/%y")
yield_curve_rates["Year"]  = yield_curve_rates["Date"].dt.year
yield_curve_rates["Month"] = yield_curve_rates["Date"].dt.month
yield_curve_rates["Day"]   = yield_curve_rates["Date"].dt.day
yield_curve_rates = yield_curve_rates.drop("Date", axis=1)
```

```python
yield_curve_rates
```

```python
bins   = [0, 3, 6, 8, 12]
labels = ["Q1", "Q2", "Q3", "Q4"]
yield_curve_rates["Quarter"] = pd.cut(yield_curve_rates["Month"], bins=bins, labels=labels, right=True)
yield_curve_rates
```
