```python
gold = pd.read_csv(f"../../../{CSV_DATA_PATH}gold_commodity.csv")
gold
```

```python
gold["Date"]  = pd.to_datetime(gold["Date"], format="%m/%d/%Y")
gold["Year"]  = gold["Date"].dt.year
gold["Month"] = gold["Date"].dt.month
gold["Day"]   = gold["Date"].dt.day
gold = gold.drop("Date", axis=1)
```

```python
gold
```

```python
bins   = [0, 3, 6, 8, 12]
labels = ["Q1", "Q2", "Q3", "Q4"]
gold["Quarter"] = pd.cut(gold["Month"], bins=bins, labels=labels, right=True)
gold
```
