```python
sp500 = pd.read_csv(f"../../../{CSV_DATA_PATH}sp500.csv")
sp500
```

```python
sp500["Date"] = pd.to_datetime(sp500["Date"], errors="coerce", format="%m/%d/%Y")
sp500["Year"] = sp500["Date"].dt.year
sp500["Month"] = sp500["Date"].dt.month
sp500["Day"] = sp500["Date"].dt.day
sp500 = sp500.drop("Date", axis=1)
```

```python
sp500
```

```python
bins   = [0, 3, 6, 8, 12]
labels = ["Q1", "Q2", "Q3", "Q4"]
sp500["Quarter"] = pd.cut(sp500["Month"], bins=bins, labels=labels, right=True)
sp500
```

```python
path = Path(Path.cwd()/"../../../../data")
path
```

```python
data_folder = path/'parquet/sp500'
data_folder.mkdir(parents=True, exist_ok=True)
```

```python
sp500.to_parquet(data_folder/"sp500.parquet")
```
