```python
dow = pd.read_csv(f"../../../{CSV_DATA_PATH}dow_jones.csv")
dow
```

```python
dow["Date"] = pd.to_datetime(dow["Date"], errors="coerce", format="%m/%d/%Y")
dow["Year"] = dow["Date"].dt.year
dow["Month"] = dow["Date"].dt.month
dow["Day"] = dow["Date"].dt.day
dow = dow.drop("Date", axis=1)
```

```python
dow
```

```python
bins   = [0, 3, 6, 8, 12]
labels = ["Q1", "Q2", "Q3", "Q4"]
dow["Quarter"] = pd.cut(dow["Month"], bins=bins, labels=labels, right=True)
dow
```

```python
path = Path(Path.cwd()/"../../../../data")
path
```

```python
data_folder = path/'parquet/dow_jones'
data_folder.mkdir(parents=True, exist_ok=True)
```

```python
dow.to_parquet(data_folder/"dow_jones.parquet")
```
