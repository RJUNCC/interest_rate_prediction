```python
nasdaq = pd.read_csv(f"../../../{CSV_DATA_PATH}nasdaq.csv")
nasdaq
```

```python
nasdaq["Date"] = pd.to_datetime(nasdaq["Date"], errors="coerce", format="%m/%d/%Y")
nasdaq["Year"] = nasdaq["Date"].dt.year
nasdaq["Month"] = nasdaq["Date"].dt.month
nasdaq["Day"] = nasdaq["Date"].dt.day
nasdaq = nasdaq.drop("Date", axis=1)
```

```python
nasdaq
```

```python
bins   = [0, 3, 6, 8, 12]
labels = ["Q1", "Q2", "Q3", "Q4"]
nasdaq["Quarter"] = pd.cut(nasdaq["Month"], bins=bins, labels=labels, right=True)
nasdaq
```

```python
path = Path(Path.cwd()/"../../../../data")
path
```

```python
data_folder = path/'parquet/nasdaq'
data_folder.mkdir(parents=True, exist_ok=True)
```

```python
nasdaq.to_parquet(data_folder/"nasdaq.parquet")
```
