```python
bill_rates = pd.read_csv(f"../../../{CSV_DATA_PATH}bill_rates.csv")
bill_rates
```

```python
bill_rates["Date"]  = pd.to_datetime(bill_rates["Date"], format="%m/%d/%y")
bill_rates["Year"]  = bill_rates["Date"].dt.year
bill_rates["Month"] = bill_rates["Date"].dt.month
bill_rates["Day"]   = bill_rates["Date"].dt.day
bill_rates = bill_rates.drop("Date", axis=1)
```

```python
bill_rates
```

```python
bins   = [0, 3, 6, 8, 12]
labels = ["Q1", "Q2", "Q3", "Q4"]
bill_rates["Quarter"] = pd.cut(bill_rates["Month"], bins=bins, labels=labels, right=True)
bill_rates
```

```python
path = Path(Path.cwd()/"../../../../data")
path
```

```python
bill_rates.to_parquet(data_folder/"bill_rates.parquet")
```
