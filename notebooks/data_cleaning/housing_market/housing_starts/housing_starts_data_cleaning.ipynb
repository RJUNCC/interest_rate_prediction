```python
starts = pd.read_csv(f"../../../{CSV_DATA_PATH}housing_starts.csv")
starts
```

```python
starts["Date"]  = pd.to_datetime(starts["Date"], format="%b %Y")
starts["Year"]  = starts["Date"].dt.year
starts["Month"] = starts["Date"].dt.month
# starts["Day"]   = starts["Date"].dt.day
starts = starts.drop("Date", axis=1)
```

```python
starts
```

```python
bins   = [0, 3, 6, 8, 12]
labels = ["Q1", "Q2", "Q3", "Q4"]
starts["Quarter"] = pd.cut(starts["Month"], bins=bins, labels=labels, right=True)
starts
```

```python
path = Path(Path.cwd()/"../../../../data")
path
```

```python
starts.to_parquet(data_folder/"housing_starts.parquet")
```
