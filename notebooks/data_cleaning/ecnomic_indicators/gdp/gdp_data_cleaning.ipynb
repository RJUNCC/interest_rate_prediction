```python
gdp = pd.read_excel(f"../../../{GDP_PATH+XL_FILE}")
gdp
```

```python
gdp = gdp.drop("Line", axis=1)
gdp
```

```python
new_name = {"Unnamed: 1":"Category"}
gdp = gdp.rename(columns=new_name)
gdp['Category'] = gdp['Category'].fillna("Quarter")
gdp
```

```python
gdp_melted = gdp.melt(id_vars=["Category"], var_name="Year", value_name="GDP")
gdp_melted
```

```python
# gdp_melted["Year"] = gdp_melted['Year'].apply(lambda x: str(x).split('.')[0])
# gdp_melted
```

```python
gdp_agg = gdp_melted.groupby(['Year', 'Category'], as_index=False).sum()
gdp_agg
```

```python
gdp_pivot = gdp_agg.pivot(index="Year", columns="Category", values="GDP")
gdp_pivot
```

```python
gdp_pivot = gdp_pivot.reset_index()
gdp_pivot
```

```python
gdp_pivot['Year'] = gdp_pivot['Year'].apply(lambda x: str(x).split('.')[0])
gdp_pivot
```

```python
gdp_pivot.columns.name = None
```

```python
gdp_pivot.columns = [col.strip() for col in gdp_pivot.columns]
```

```python
path = Path(Path.cwd()/"../../../../data")
path
```

```python
data_folder = path/'parquet/gdp'
data_folder.mkdir(parents=True, exist_ok=True)
```

```python
gdp_pivot.columns
```

```python
def make_unique(column_names):
    seen = {}
    result = []
    for col in column_names:
        if col not in seen:
            seen[col] = 1
            result.append(col)
        else:
            seen[col] += 1
            result.append(f"{col}_{seen[col]}")
    return result
```

```python
gdp_pivot.columns = make_unique(gdp_pivot.columns)
```

```python
gdp_pivot.isna().sum()
```

```python
gdp_pivot.to_parquet(data_folder/"gdp.parquet", index=False)
```
