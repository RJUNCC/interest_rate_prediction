```python
all_items = pd.read_excel(f"../../../{CPI_PATH}allitems.xlsx").T.reset_index()
all_items.head()
```

```python
all_items.columns
```

```python
all_items = all_items.drop(all_items.columns[:11], axis=1)
all_items
```

```python
all_items = all_items.set_index(keys=10)
all_items
```

```python
all_items = all_items.T.drop(['HALF1', 'HALF2'], axis=1)
all_items
```

```python
all_items_melted = all_items.melt(id_vars=['Year'], var_name="Month", value_name="CPI")
all_items_melted = all_items_melted.sort_values(by='Year')
all_items_melted = all_items_melted.reset_index(drop=True)
all_items_melted
```

```python
path = Path(Path.cwd()/"../../../../data")
path
```

```python
data_folder = path/'parquet/cpi'
data_folder.mkdir(parents=True, exist_ok=True)
```
