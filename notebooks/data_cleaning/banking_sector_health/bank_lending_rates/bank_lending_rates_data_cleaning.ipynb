```python
bank_rates = pd.read_csv(f"../../../{CSV_DATA_PATH}PRIME.csv")
bank_rates
```

```python
bank_rates["DATE"]  = pd.to_datetime(bank_rates["DATE"], format="%Y-%m-%d")
bank_rates["Year"]  = bank_rates["DATE"].dt.year
bank_rates["Month"] = bank_rates["DATE"].dt.month
bank_rates["Day"]   = bank_rates["DATE"].dt.day
bank_rates = bank_rates.drop("DATE", axis=1)
```

```python
bank_rates = bank_rates.sort_values(["Year", "Month", "Day"])
```

```python
bank_rates
```

```python
bins   = [0, 3, 6, 8, 12]
labels = ["Q1", "Q2", "Q3", "Q4"]
bank_rates["Quarter"] = pd.cut(bank_rates["Month"], bins=bins, labels=labels, right=True)
bank_rates
```
