```python
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
def rename_duplicate_columns(df):
    cols = pd.Series(df.columns)
    for dup in cols[cols.duplicated()].unique():
        cols[cols[cols == dup].index.values.tolist()] = [dup + '_' + str(i) if i != 0 else dup for i in range(sum(cols == dup))]
    df.columns = cols
    return df
```

```python
merged = rename_duplicate_columns(merged)
```

```python
bad_columns = ['8 WEEKS BANK DISCOUNT', '8 WEEKS COUPON EQUIVALENT',
 '17 WEEKS BANK DISCOUNT', '17 WEEKS COUPON EQUIVALENT', '20 YR', '30 YR',
 '2 Mo', '4 Mo', 'Close/Last_nsdq', 'Volume', 'Open_dow', 'High_dow', 'Low_dow',
 'Close/Last_nsdq', 'Open_sp500', 'High_sp500', 'Low_sp500', 'PPI',
 'Unemployment_Rate']

for col in bad_columns:
    temp_mean = merged[col].mean()
    merged[col] = merged[col].fillna(temp_mean)
```

```python
very_bad_columns = ['52 WEEKS BANK DISCOUNT', '52 WEEKS COUPON EQUIVALENT', '30 Yr', 'Close/Last_nsdq_1', '5 YR', '7 YR', '10 YR']
merged = merged.drop(columns=very_bad_columns, axis=1)
```

```python
exclude_columns = ["Year", "Month", "Day", "Quarter"]
```

```python
target_variable = "Fed_Fund_rate"
```

```python
numeric_columns = [col for col in merged.select_dtypes(include=[np.number]).columns if col not in exclude_columns + [target_variable]]
categorical_columns = [col for col in merged.select_dtypes(include=['object']).columns if col not in exclude_columns]
```

```python
class ChunkMeanImputer(BaseEstimator, TransformerMixin):
    def __init__(self, chunk_size=100):
        self.chunk_size = chunk_size
        self.imputers_ = []
    
    def fit(self, X, y=None):
        self.imputers_ = []
        n_chunks = int(np.ceil(X.shape[0] / self.chunk_size))
        
        for i in range(n_chunks):
            chunk = X[i * self.chunk_size: (i + 1) * self.chunk_size]
            imputer = SimpleImputer(strategy='mean')
            imputer.fit(chunk)
            self.imputers_.append(imputer)
        
        return self
    
    def transform(self, X):
        X_transformed = np.empty_like(X)
        
        for i, imputer in enumerate(self.imputers_):
            chunk = X[i * self.chunk_size: (i + 1) * self.chunk_size]
            X_transformed[i * self.chunk_size: (i + 1) * self.chunk_size] = imputer.transform(chunk)
        
        return X_transformed
```

```python
numeric_transformer = make_pipeline(
    ChunkMeanImputer(chunk_size=100),
    StandardScaler()
)
```

```python
categorical_transformer = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown='ignore')
)
```

```python
preprocessor = make_column_transformer(
    (numeric_transformer, numeric_columns),
    (categorical_transformer, categorical_columns)
)
```

```python
X = merged.drop(columns=["Fed_Fund_rate"], axis=1)
y = merged['Fed_Fund_rate']
```

```python
excluded_columns_df = merged[exclude_columns]
```

```python
X_transformed = preprocessor.fit_transform(X)
```

```python
X_transformed
```

```python
numeric_feature_names = numeric_columns
categorical_feature_names = preprocessor.transformers_[1][1][1].get_feature_names_out(categorical_columns).tolist()
transformed_columns = numeric_feature_names + categorical_feature_names
```

```python
# Convert the transformed array back to a DataFrame
X_transformed_df = pd.DataFrame(X_transformed.toarray(), columns=transformed_columns, index=X.index)
X_transformed_df
```

```python
target_variable = "Fed_Fund_rate"
```

```python
# Re-add the excluded columns
X_transformed_df[exclude_columns] = excluded_columns_df

# Ensure target variable is included for the year 2024 processing
df_transformed = X_transformed_df.copy()
df_transformed[target_variable] = y
```

```python
# Separate rows where Year is 2024
X_2024 = df_transformed[df_transformed['Year'] == 2024].drop(columns=[target_variable]+exclude_columns)
y_2024 = df_transformed[df_transformed['Year'] == 2024][target_variable]

# Separate rows where Year is not 2024
X_not_2024 = df_transformed[df_transformed['Year'] != 2024].drop(columns=[target_variable]+exclude_columns)
y_not_2024 = df_transformed[df_transformed['Year'] != 2024][target_variable]

# Perform a 70/30 split on the features and target for training and testing
X_train, X_test, y_train, y_test = train_test_split(X_not_2024, y_not_2024, test_size=0.30, random_state=42)

# Train a model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict values for 2024 data
predictions_2024 = model.predict(X_2024)

# Fill missing values in y_2024 with the predictions
y_2024_filled = y_2024.fillna(pd.Series(predictions_2024, index=y_2024.index))

# Combine the filled 2024 data back with the rest of the data
y_combined = pd.concat([y_not_2024, y_2024_filled])

# Reconstruct the DataFrame with the filled target variable
df_filled = X_transformed_df.copy()
df_filled[target_variable] = y_combined
```

```python
# Reorder columns so excluded columns come first
df_filled = df_filled[exclude_columns + transformed_columns + [target_variable]]
df_filled
```

```python
# Perform a 70/30 split on the filled DataFrame
train_df, test_df = train_test_split(df_filled, test_size=0.30, random_state=42)

# # Display the shapes of the resulting DataFrames
# print("Training DataFrame shape:", train_df.shape)
# print("Testing DataFrame shape:", test_df.shape)
```
